```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
activations = ["relu", "sigmoid", "tanh", "linear", "leakyrelu", "gelu"]
base_path = "/mnt/d/CUDA/day26"

relu_fp64 = np.fromfile(os.path.join(base_path, "relu_fp64.bin"), dtype=np.float64)
relu_int8 = np.fromfile(os.path.join(base_path, "relu_int8.bin"), dtype=np.int8)

sigmoid_fp64 = np.fromfile(os.path.join(base_path, "sigmoid_fp64.bin"), dtype=np.float64)
sigmoid_int8 = np.fromfile(os.path.join(base_path, "sigmoid_int8.bin"), dtype=np.int8)

tanh_fp64 = np.fromfile(os.path.join(base_path, "tanh_fp64.bin"), dtype=np.float64)
tanh_int8 = np.fromfile(os.path.join(base_path, "tanh_int8.bin"), dtype=np.int8)

linear_fp64 = np.fromfile(os.path.join(base_path, "linear_fp64.bin"), dtype=np.float64)
linear_int8 = np.fromfile(os.path.join(base_path, "linear_int8.bin"), dtype=np.int8)

leakyrelu_fp64 = np.fromfile(os.path.join(base_path, "leakyrelu_fp64.bin"), dtype=np.float64)
leakyrelu_int8 = np.fromfile(os.path.join(base_path, "leakyrelu_int8.bin"), dtype=np.int8)

gelu_fp64 = np.fromfile(os.path.join(base_path, "gelu_fp64.bin"), dtype=np.float64)
gelu_int8 = np.fromfile(os.path.join(base_path, "gelu_int8.bin"), dtype=np.int8)
```

```python
fig, axs = plt.subplots(2, 6, figsize=(24, 8))

axs[0, 0].hist(relu_fp64, bins=50, color='blue', alpha=0.7, edgecolor='black')
axs[0, 0].set_title("ReLU (FP64)")
axs[1, 0].hist(relu_int8, bins=50, color='red', alpha=0.7, edgecolor='black')
axs[1, 0].set_title("ReLU (INT8)")

axs[0, 1].hist(sigmoid_fp64, bins=50, color='blue', alpha=0.7, edgecolor='black')
axs[0, 1].set_title("Sigmoid (FP64)")
axs[1, 1].hist(sigmoid_int8, bins=50, color='red', alpha=0.7, edgecolor='black')
axs[1, 1].set_title("Sigmoid (INT8)")

axs[0, 2].hist(tanh_fp64, bins=50, color='blue', alpha=0.7, edgecolor='black')
axs[0, 2].set_title("Tanh (FP64)")
axs[1, 2].hist(tanh_int8, bins=50, color='red', alpha=0.7, edgecolor='black')
axs[1, 2].set_title("Tanh (INT8)")

axs[0, 3].hist(linear_fp64, bins=50, color='blue', alpha=0.7, edgecolor='black')
axs[0, 3].set_title("Linear (FP64)")
axs[1, 3].hist(linear_int8, bins=50, color='red', alpha=0.7, edgecolor='black')
axs[1, 3].set_title("Linear (INT8)")

axs[0, 4].hist(leakyrelu_fp64, bins=50, color='blue', alpha=0.7, edgecolor='black')
axs[0, 4].set_title("LeakyReLU (FP64)")
axs[1, 4].hist(leakyrelu_int8, bins=50, color='red', alpha=0.7, edgecolor='black')
axs[1, 4].set_title("LeakyReLU (INT8)")

axs[0, 5].hist(gelu_fp64, bins=50, color='blue', alpha=0.7, edgecolor='black')
axs[0, 5].set_title("GELU (FP64)")
axs[1, 5].hist(gelu_int8, bins=50, color='red', alpha=0.7, edgecolor='black')
axs[1, 5].set_title("GELU (INT8)")

for i in range(6):
    axs[1, i].set_xlabel("Activation Value")
    axs[0, i].set_ylabel("Frequency")
    axs[1, i].set_ylabel("Frequency")

fig.suptitle("Activation Distributions: FP64 vs. INT8", fontsize=16, fontweight='bold')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
